```python
import pandas as pd
```

```python
section = pd.read_csv('network_data/result_section.csv', header=0)
cable = pd.read_csv('network_data/cable_combine.csv', header=0)
section = section[['SectionID', 'FromNodeID', 'ToNodeID']]
section= section.applymap(str)
cable['SectionID'] = cable['SectionID'].astype(str)

cable_section = cable.merge(section, on='SectionID')
cable_section['B1'] = cable_section['B1'].astype(float)

#with pd.option_context('display.max_rows', None, 'display.max_columns', None):
#    print(cable_section)
```

```python
# Checking R/X of the lines in order to check the rated voltage of each bus
## R/X big than low voltage (0.42)

cable_section_2 = cable_section[cable_section['X1'] >= 0.001].reset_index()

cable_section_2['RX'] = cable_section_2['R1'] / cable_section_2['X1']

#cable_section_2 = cable_section_2[cable_section_2['SectionID'] == '42600040']

#with pd.option_context('display.max_rows', None, 'display.max_columns', None):
#    print(cable_section_2)

#voltage = cable_section_2[cable_section_2['R/X'] < 1]    
#voltage
cable_section_2 = cable_section[cable_section['X1'] >= 0.001].reset_index()

cable_section_2['RX'] = cable_section_2['R1'] / cable_section_2['X1']
r_voltage = []

for row in cable_section_2['RX']:
    if 0 <= row  <= 1:
        r_voltage.append(10.6) 
    if row > 1:
        r_voltage.append(0.42)
        
cable_section_2['r_voltage'] = r_voltage

cable_section_2
```
