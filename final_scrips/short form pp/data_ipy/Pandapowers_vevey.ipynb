```python
import pandapower as pp
import pandapower.networks as ppnx
import pandas as pd
import json
import pandapower.networks as nw
import matplotlib.pyplot as plt
import pandapower.plotting as plot
```

```python
my_net = nw.pp.create_empty_network()
```

```python
bus = json.load(open('bus.json'))
my_net.bus = pd.read_json(bus)
```

```python
line = json.load(open('line.json'))
my_net.line = pd.read_json(line)
```

```python
switch = json.load(open('switch.json'))
my_net.switch = pd.read_json(switch)
```

```python
load = json.load(open('load.json'))
my_net.load = pd.read_json(load)
```

```python
ext_grid = json.load(open('ext_grid.json'))
my_net.ext_grid = pd.read_json(ext_grid)
```

```python
sgen = json.load(open('sgen.json'))
my_net.sgen = pd.read_json(sgen)
```

```python
trafo = json.load(open('trafo.json'))
my_net.trafo = pd.read_json(trafo)
```

```python
bus_geodata = json.load(open('bus_geodata.json'))
my_net.bus_geodata = pd.read_json(bus_geodata)
```

```python
line_geodata = json.load(open('line_geodata.json'))
my_net.line_geodata = pd.read_json(line_geodata)
```

```python
pp.runpp(my_net, max_iteration=10000)
```

```python
cmap_list=[(20, "green"), (50, "yellow"), (60, "red")]
cmap, norm = plot.cmap_continous(cmap_list)
lc = plot.create_line_collection(my_net, my_net.line.index, zorder=1, cmap=cmap, norm=norm, linewidths=5)

cmap_list=[(0, "blue"), (1.0, "green"), (1.1, "red")]
cmap, norm = plot.cmap_continous(cmap_list)
bc = plot.create_bus_collection(my_net, my_net.bus.index, size=5, zorder=2, cmap=cmap, norm=norm)

loadc = plot.create_bus_collection(my_net, my_net.sgen.index, size=5, zorder=1, norm=norm)

plot.draw_collections([lc, bc], figsize=(18,10))
```
