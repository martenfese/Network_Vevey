```python
import pandas as pd
```

```python
coord = pd.read_csv('../preprocesses/section_data/net_coord.csv')
coord = coord[["NodeID","CoordX", "CoordY"]].copy()

coord = coord[coord["CoordX"] <= 555000]

coord
```

```python
subnetwork_connections = pd.read_csv('../preprocesses/section_data/result_subnetwork_connections.csv')

subnetwork_connections = subnetwork_connections[["NodeID", "ConnectorCoordX", "ConnectorCoordY"]].copy()

subnetwork_connections = subnetwork_connections.rename(index=str, columns={"ConnectorCoordX": "CoordX", "ConnectorCoordY": "CoordY"})

subnetwork_connections['NodeID'] = subnetwork_connections['NodeID'].astype(str)

with pd.option_context('display.max_rows', None, 'display.max_columns', 6):
    print(subnetwork_connections)
```

```python
frames = [coord,subnetwork_connections] 
coord = pd.concat(frames)

coord
```

```python
### Create a dataframe with the nodes that are in a subnetwork and have missing coordinates
# There is definitly a shorter and more sexy way, but ...

section_orig = pd.read_csv('../preprocesses/section_data/result_section_better.csv')

section = section_orig[["SectionID", "FromNodeID", "ToNodeID","SubNetworkId"]].copy()
section = section.dropna() # Rows without a subnetwork are dropped
section = section.reset_index()

section['SectionID'] = section['SectionID'].astype(str)
section['FromNodeID'] = section['FromNodeID'].astype(str)
section['ToNodeID'] = section['ToNodeID'].astype(str)
section["SubNetworkId"] = section["SubNetworkId"].astype(str)

section_from = section[["FromNodeID", "SubNetworkId"]]
section_to = section[["ToNodeID", "SubNetworkId"]]
section_to = section_to[~section_to['ToNodeID'].isin(coord['NodeID'])].reset_index() # only nodes where coordiantes are missing
section_to  = section_to[["ToNodeID","SubNetworkId"]].copy() # get rid of the old index...
section_to  # contains all nodes from ToNode with no coordinates

section_from = section[["FromNodeID", "SubNetworkId"]]
section_from = section_from[~section_from['FromNodeID'].isin(coord['NodeID'])].reset_index() # only nodes where coordiantes are missing
section_from  = section_from[["FromNodeID","SubNetworkId"]].copy() # get rid of the old index...
section_from  # contains all nodes from FromNode with no coordinates

#section_both = section_from[~section_from['FromNodeID'].isin(section_to['ToNodeID'])]
section_to = section_to.rename(index=str, columns={"ToNodeID": "NodeID"})
section_from = section_from.rename(index=str, columns={"FromNodeID": "NodeID"})


frames = [section_to, section_from] 
section_both = pd.concat(frames)

section_both # contains all nodes with missing coordinates that are part of a Subnetwork

section_both = section_both.drop_duplicates(subset="NodeID").reset_index()
section_both = section_both[["NodeID", "SubNetworkId"]].copy()


with pd.option_context('display.max_rows', None, 'display.max_columns', 6):
    print(section_both)
```

```python
subnetworks = pd.read_csv('../preprocesses/section_data/subnetworks.csv')

subnetworks = subnetworks[["SubNetID", "X", "Y"]].copy()

subnetworks = subnetworks.rename(index=str, columns={"SubNetID": "SubNetworkId","X":"CoordX","Y":"CoordY"})

subnetworks_as_nodes = subnetworks.rename(index=str, columns={"SubNetworkId": "NodeID","X":"CoordX","Y":"CoordY"})

subnetworks['SubNetworkId'] = subnetworks['SubNetworkId'].astype(str)

subnetworks
```

```python
sub_network_nodes = pd.DataFrame(columns=["NodeID","CoordX", "CoordY", "SubNetworkId"])
section_dict = {}


for j in range(1432):
    for i in range(101):
        if section_both["SubNetworkId"][j] == subnetworks['SubNetworkId'][i]:
            section_dict.update({'NodeID': section_both['NodeID'][j],
                                 'CoordX': subnetworks['CoordX'][i], 
                                 'CoordY': subnetworks['CoordY'][i],
                                 'SubNetworkId': subnetworks['SubNetworkId'][i]})
            sub_network_nodes = sub_network_nodes.append(section_dict.copy(),ignore_index=True)

#sub_network_nodes.to_csv(r'subnetworks_nodes_new.csv', header=True, index=True, mode='a')
```
